Exploratory Data Analysis (EDA) adalah sebuah proses kritis dalam melakukan investigasi awal pada data dengan tujuan menemukan pola, anomali, menguji hipotesis dan dapat memeriksa asumsi dengan bantuan statistik ringkasan kemudian representasi grafis (visualisasi).


Kaidah filosofi dari Exploratory Data Analysis (EDA) adalah sebagai berikut:

1. Menggunakan grafik untuk memeriksa data.

2. Menggunakan statistik deskriptif untuk merangkum data.

3. Menggunakan teknik-teknik matematika untuk mengidentifikasi model yang mungkin.

4. Mengeksplorasi hubungan antara variabel.

  Dalam pemilihan warna pada grafik, terdapat beberapa konsep dasar pemilihan desain warna. Berikut adalah beberapa konsep dasar pemilihan desain warna:

1. Analogous colors: Warna yang berdekatan pada roda warna.

2. Complementary colors: Warna yang berlawanan pada roda warna.

3. Triadic colors: Tiga warna yang berjarak sama pada roda warna.

4. Monochromatic colors: Warna yang berasal dari satu warna dasar.
